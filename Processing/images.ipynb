```python
import os
import pandas as pd

# Function to compare directory files with CSV file
def list_missing_files(csv_file, directory, column_name='file_name'):
    # Read the CSV file
    csv_data = pd.read_csv(csv_file)
    
    # Ensure the column exists
    if column_name not in csv_data.columns:
        raise ValueError(f"The column '{column_name}' does not exist in the CSV file.")
    
    # Get the list of filenames from the CSV file
    csv_filenames = set(csv_data[column_name].astype(str).tolist())
    
    # Get the list of filenames in the directory
    directory_filenames = set(os.listdir(directory))
    
    # Find files in the directory that are not in the CSV file
    missing_files = directory_filenames - csv_filenames
    
    return list(missing_files)

# Example usage
csv_file_path = r"D:\Documents\MASC\MSP_POD_dataset\Audios\Audios.tar\Audios\metadata.csv"  # Path to your CSV file
directory_path = r"D:\Documents\MASC\MSP_POD_dataset\Audios\Audios.tar\Audios"  # Path to your directory

missing_files = list_missing_files(csv_file_path, directory_path)

print("Files in the directory that are not listed in the CSV file:")
for file in missing_files:
    print(file)
```

```python
# Hugging Face Datasets
from datasets import load_dataset, concatenate_datasets

# Data processing and metrics
import numpy as np
from sklearn.metrics import (
    accuracy_score,
    recall_score,
    f1_score,
    confusion_matrix,
)

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, clear_output


dataset_name = 'cairocode/IEMO_WAV_002'
dataset = load_dataset(dataset_name)
from datasets import DatasetDict
from PIL import Image
import matplotlib.pyplot as plt
import io
import torch
from transformers import Wav2Vec2Processor, Wav2Vec2Model
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
import logging
from transformers import logging as transformers_logging
# Set the logging level to ERROR to suppress warnings
logging.getLogger("transformers").setLevel(logging.ERROR)
transformers_logging.set_verbosity_error()


# Updated audio_to_image function to return an image
def audio_to_image(audio_array, sample_rate=16000, model_name="facebook/wav2vec2-base-960h"):
    """
    Takes an input audio array and uses Wav2Vec2 to generate an image representation.

    Args:
        audio_array (numpy array): Input audio data as a 1D array.
        sample_rate (int): Sampling rate of the audio.
        model_name (str): Pretrained Wav2Vec2 model name.

    Returns:
        PIL.Image: Image representation of the audio features.
    """
    processor = Wav2Vec2Processor.from_pretrained(model_name)
    model = Wav2Vec2Model.from_pretrained(model_name)

    inputs = processor(audio_array, sampling_rate=sample_rate, return_tensors="pt", padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
 
    hidden_states = outputs.last_hidden_state.squeeze(0).numpy()

    plt.figure(figsize=(10, 6))
    plt.imshow(hidden_states.T, aspect="auto", origin="lower", cmap="viridis")
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    plt.close()
    buf.seek(0)

    image = Image.open(buf).convert('RGB')
    image = image.resize((224, 224))  # Resize to 224x224
    buf.close()
    return image

# Function to process a batch
def process_batch(batch, sample_rate=16000, model_name="facebook/wav2vec2-base-960h"):
    images = []
    for audio_data in tqdm(batch['audio'], desc="Processing batch"):
        image = audio_to_image(audio_data['array'], sample_rate=sample_rate, model_name=model_name)
        images.append(image)
    return {'image': images}



```

```python
import os
from pathlib import Path
import wave
import numpy as np
from tqdm import tqdm
from PIL import Image
import matplotlib.pyplot as plt
from transformers import Wav2Vec2Processor, Wav2Vec2Model
import torch
import io
import warnings
import pandas as pd
import logging
from transformers import logging as transformers_logging
# Set the logging level to ERROR to suppress warnings
logging.getLogger("transformers").setLevel(logging.ERROR)
transformers_logging.set_verbosity_error()

warnings.filterwarnings("ignore", category=UserWarning)

# Function to convert audio to image
def audio_to_image(audio_array, sample_rate=16000, model_name="facebook/wav2vec2-base-960h"):
    processor = Wav2Vec2Processor.from_pretrained(model_name)
    model = Wav2Vec2Model.from_pretrained(model_name)

    # Convert to float32
    inputs = processor(audio_array.astype(np.float32), sampling_rate=sample_rate, return_tensors="pt", padding=True)
    with torch.no_grad():
        outputs = model(**inputs)

    hidden_states = outputs.last_hidden_state.squeeze(0).numpy()

    plt.figure(figsize=(10, 6))
    plt.imshow(hidden_states.T, aspect="auto", origin="lower", cmap="viridis")
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    plt.close()
    buf.seek(0)

    image = Image.open(buf).convert('RGB')
    image = image.resize((224, 224))  # Resize to 224x224
    buf.close()
    return image


# Function to process a directory of .wav files in batches
def process_directory_in_batches(wav_dir, output_dir, batch_size=10, sample_rate=16000, model_name="facebook/wav2vec2-base-960h"):
    wav_files = list(Path(wav_dir).glob("*.wav"))
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    # Process files in batches
    for i in tqdm(range(0, len(wav_files), batch_size), desc="Processing batches"):
        batch_files = wav_files[i:i + batch_size]
        images = []

        for wav_file in batch_files:
            # Read .wav file
            with wave.open(str(wav_file), "rb") as wf:
                n_channels = wf.getnchannels()
                n_frames = wf.getnframes()
                audio_data = wf.readframes(n_frames)
                audio_array = np.frombuffer(audio_data, dtype=np.int16)
                if n_channels > 1:
                    audio_array = audio_array[::n_channels]  # Convert to mono

            # Convert to image
            image = audio_to_image(audio_array, sample_rate=sample_rate, model_name=model_name)
            images.append((image, wav_file.stem))

        # Save images
        for image, stem in images:
            output_file = output_dir / f"{stem}.png"
            image.save(output_file, "PNG")

# Example usage
wav_directory = r"D:\Documents\MASC\MSP_POD_dataset\Audios\Audios.tar\Audios"  # Replace with your directory containing .wav files
output_directory = r"D:\Documents\MASC\MSP_POD_dataset\Image_DS\train"  # Replace with your desired output directory

process_directory_in_batches(wav_directory, output_directory, batch_size=10)


input_csv = os.path.join(wav_directory, "metadata.csv")
df = pd.read_csv(input_csv)

# (Optionally update the file_name in the DataFrame to .txt)
df["file_name"] = df["file_name"].str.replace(".wav", ".png", regex=False)

# Write to a new CSV
output_csv = os.path.join(output_directory, "metadata.csv")
df.to_csv(output_csv, index=False)
```

```python
import os
from pathlib import Path
import wave
import numpy as np
from tqdm import tqdm
from PIL import Image
import matplotlib.pyplot as plt
from transformers import Wav2Vec2Processor, Wav2Vec2Model
import torch
import io
import warnings
import pandas as pd
import logging
from transformers import logging as transformers_logging

# Set the logging level to ERROR to suppress warnings
logging.getLogger("transformers").setLevel(logging.ERROR)
transformers_logging.set_verbosity_error()

warnings.filterwarnings("ignore", category=UserWarning)
model_name="facebook/wav2vec2-base-960h"
processor = Wav2Vec2Processor.from_pretrained(model_name)
model = Wav2Vec2Model.from_pretrained(model_name)


# Function to convert audio to image
def audio_to_image(audio_array, sample_rate=16000):
    
    # Convert to float32
    inputs = processor(audio_array.astype(np.float32), sampling_rate=sample_rate, return_tensors="pt", padding=True)
    with torch.no_grad():
        outputs = model(**inputs)

    hidden_states = outputs.last_hidden_state.squeeze(0).numpy()

    plt.figure(figsize=(10, 6))
    plt.imshow(hidden_states.T, aspect="auto", origin="lower", cmap="viridis")
    plt.axis('off')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    plt.close()
    buf.seek(0)

    image = Image.open(buf).convert('RGB')
    image = image.resize((224, 224))  # Resize to 224x224
    buf.close()
    return image

# Function to process a directory of .wav files in batches
def process_directory_in_batches(wav_dir, output_dir, batch_size=20, sample_rate=16000):
    wav_files = list(Path(wav_dir).glob("*.wav"))
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)
    print("Length of Wav files: ", len(wav_files))

    # Precompute existing output files
    existing_files = {f.stem for f in output_dir.glob("*.png")}
    print("Length of Existing files: ", len(existing_files))
    wav_files_to_process = [wav for wav in wav_files if wav.stem not in existing_files]
    print("Length of Wav files to prrocess: ", len(wav_files_to_process))

    # Process files in batches
    for i in tqdm(range(0, len(wav_files_to_process), batch_size), desc="Processing batches"):
        batch_files = wav_files_to_process[i:i + batch_size]
        images = []

        for wav_file in batch_files:
            # Read .wav file
            with wave.open(str(wav_file), "rb") as wf:
                n_channels = wf.getnchannels()
                n_frames = wf.getnframes()
                audio_data = wf.readframes(n_frames)
                audio_array = np.frombuffer(audio_data, dtype=np.int16)
                if n_channels > 1:
                    audio_array = audio_array[::n_channels]  # Convert to mono

            # Convert to image
            image = audio_to_image(audio_array, sample_rate=sample_rate)
            images.append((image, wav_file.stem))

        # Save images
        for image, stem in images:
            output_file = output_dir / f"{stem}.png"
            image.save(output_file, "PNG")

# Example usage
wav_directory = r"D:\Documents\MASC\MSP_POD_dataset\Audios\Audios.tar\Audios"  # Replace with your directory containing .wav files
output_directory = r"D:\Documents\MASC\MSP_POD_dataset\Image_DS\train"  # Replace with your desired output directory

process_directory_in_batches(wav_directory, output_directory, batch_size=20)

input_csv = os.path.join(wav_directory, "metadata.csv")
df = pd.read_csv(input_csv)

# (Optionally update the file_name in the DataFrame to .txt)
df["file_name"] = df["file_name"].str.replace(".wav", ".png", regex=False)

# Write to a new CSV
output_csv = os.path.join(output_directory, "metadata.csv")
df.to_csv(output_csv, index=False)
```
